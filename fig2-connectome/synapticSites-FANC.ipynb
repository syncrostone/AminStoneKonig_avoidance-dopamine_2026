```python
# https://github.com/htem/FANC_auto_recon for information on api key. Paste yours below or in ./data/fanc_api_key.txt 
api_key = "your api key here"
if api_key == "your api key here":
    with open("../data/fanc_api_key.txt", "r", encoding="utf-8") as f:
        api_key = f.read().strip()
fanc.save_cave_credentials(api_key, overwrite=True)
```

```python
# The following is a simpler alternative to `client = CAVEclient('fanc_production_mar2021')`
client = fanc.get_caveclient()
```

```python
mdn_segids = fanc.lookup.cells_annotated_with('MDN')
segid = mdn_segids[1] # Select one of the MDN neurons to visualize (1,2,3,4)

outputs = client.materialize.synapse_query(pre_ids=segid)  # after clicking the link, make sure to color the outputs differently from the inputs, so we can distinguish them in the visualization
inputs = client.materialize.synapse_query(post_ids=segid)

fanc.statebuilder.render_scene(neurons=segid, annotations=[{'type': 'points', 'name': 'inputs', 'data': inputs.pre_pt_position.values}, {'type': 'points', 'name': 'outputs', 'data': outputs.post_pt_position.values}], bg_color='w')
```
