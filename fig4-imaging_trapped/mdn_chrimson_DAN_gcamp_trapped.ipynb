# Fig 4e

```python
dat = pd.read_csv('stats_trapped_imaging.csv')
```

```python
pg.rm_anova(data=dat, subject='fly', within=['trapping', 'compartment'], dv='deltaF')
```

```python
fig, axs = plt.subplots(2, 3, sharey=True, figsize=(10,8))
trapping_order = ['pre', 'trapped', 'post-trapping']
pg.plot_paired(data=dat.loc[dat.compartment=='gamma1'], subject='fly', within='trapping', 
                  dv='deltaF', order=trapping_order, boxplot_kwargs={'whis':[10,90]},
                  ax=axs[0,0]).set_title('gamma1')
pg.plot_paired(data=dat.loc[dat.compartment=='gamma2'], subject='fly', within='trapping', 
                  dv='deltaF', order=trapping_order, boxplot_kwargs={'whis':[10,90]},
                  ax=axs[0,1]).set_title('gamma2')
pg.plot_paired(data=dat.loc[dat.compartment=='gamma3'], subject='fly', within='trapping', 
                  dv='deltaF', order=trapping_order, boxplot_kwargs={'whis':[10,90]},
                  ax=axs[0,2]).set_title('gamma3')
pg.plot_paired(data=dat.loc[dat.compartment=='gamma4'], subject='fly', within='trapping', 
                  dv='deltaF', order=trapping_order, boxplot_kwargs={'whis':[10,90]},
                  ax=axs[1,0]).set_title('gamma4')
pg.plot_paired(data=dat.loc[dat.compartment=='gamma5'], subject='fly', within='trapping', 
                  dv='deltaF', order=trapping_order, boxplot_kwargs={'whis':[10,90]},
                  ax=axs[1,1]).set_title('gamma5')
pg.plot_paired(data=dat.loc[dat.compartment=='betap2'], subject='fly', within='trapping', 
                  dv='deltaF', order=trapping_order, boxplot_kwargs={'whis':[10,90]},
                  ax=axs[1,2]).set_title('betap2')
for ax in axs.flatten():
    ax.set_xticklabels(trapping_order, rotation=45)
fig.tight_layout()
```
