```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import pingouin as pg
```

# Ext Data Fig 6b

```python
responses_df = pd.read_csv('gamma3_responses_specific.csv')
```

```python
pg.normality(responses_df.calcium_response)
```

```python
pg.friedman(data=responses_df, dv='calcium_response', within='ML_position', subject='index')
```

```python
pg.pairwise_tests(data=responses_df, dv='calcium_response', within='ML_position', subject='index', parametric=False,
                  padjust='holm')
```

```python
responses_df.ML_position = responses_df.ML_position.astype(int)
```

```python
sns.regplot(data=responses_df, x='ML_position', y='calcium_response')
sns.lineplot(data=responses_df, x='ML_position', y='calcium_response', estimator=None, units='index')
plt.gca().invert_xaxis()
plt.ylim([0.15, 2.1]);
```
